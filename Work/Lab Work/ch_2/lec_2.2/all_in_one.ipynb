## Q1. Discrete Random Variable – Dice Roll
- Simulate rolling a fair 6-sided dice 1000 times.
- Store the results as a discrete random variable.
- Plot the frequency of each outcome.

```python
import numpy as np
import matplotlib.pyplot as plt

dice = np.random.randint(1, 7, 1000)
values, counts = np.unique(dice, return_counts=True)

plt.bar(values, counts)
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.title('Frequency of Dice Outcomes (1000 Rolls)')
plt.show()
```

## Q2. Continuous Random Variable – Heights of Students
- Generate heights of 50 students using a normal distribution.
- Treat it as a continuous random variable.
- Plot its histogram.

```python
heights = np.random.normal(loc=165, scale=10, size=50)

plt.hist(heights, bins=10)
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Student Heights')
plt.show()
```

## Q3. Probability Distribution – Coin Toss
- Simulate 500 tosses of a fair coin.
- Calculate probability of heads and tails.
- Plot the probability distribution.

```python
coin = np.random.choice(['Heads', 'Tails'], size=500)
values, counts = np.unique(coin, return_counts=True)
probabilities = counts / 500

plt.bar(values, probabilities)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Distribution of Coin Toss')
plt.show()
```

## Q4. Probability Distribution – Rolling a Dice
- Simulate rolling a 6-sided dice 10,000 times.

## Q5. Mean – Expected Value of a Dice Roll
E[X] = Σ x · P(x)

```python
faces = np.arange(1, 7)
prob = np.ones(6) / 6

expected_value = np.sum(faces * prob)
expected_value
```

## Q6. Mean – Expected Value of Coin Toss Game
- Win $2 for Heads
- Lose $1 for Tails

```python
earnings = np.array([2, -1])
prob = np.array([0.5, 0.5])

expected_earnings = np.sum(earnings * prob)
expected_earnings
```

## Q7. Variance – Dice Roll
Var(X) = Σ (x − μ)² · P(x)

```python
mean = expected_value
variance = np.sum((faces - mean)**2 * prob)
variance
```

## Q8. Variance – Game Earnings
- Use the coin toss game from Q6.
- Calculate the variance of earnings.
- Interpret what a high or low variance means.

```python
mean_earn = expected_earnings
variance_earn = np.sum((earnings - mean_earn)**2 * prob)
variance_earn
```
