## Q1 – SVD Decomposition

```python
import numpy as np

A = np.random.rand(4,3)

U, S, VT = np.linalg.svd(A)

print('Matrix A:\n', A)
print('Singular Values:', S)

# Reconstruction using top 2 singular values
S_matrix = np.zeros((4,3))
S_matrix[:3, :3] = np.diag(S)

# Keep top 2
S_matrix[2:, :] = 0

A_approx = U @ S_matrix @ VT

print('Reconstructed Matrix (Top 2 SVD):\n', A_approx)
```

## Q2 – PCA (Manual + Sklearn)

```python
import matplotlib.pyplot as plt

# Synthetic correlated data
np.random.seed(0)
x = np.random.rand(100)
y = 2*x + np.random.normal(0, 0.1, 100)

X = np.vstack((x, y)).T

plt.scatter(x, y)
plt.title('Correlated Data')
plt.show()

# Covariance matrix
cov = np.cov(X.T)

# Eigen decomposition
eigvals, eigvecs = np.linalg.eig(cov)

print('Eigenvalues:', eigvals)
print('Eigenvectors:\n', eigvecs)

# Project on first principal component
pc1 = eigvecs[:, np.argmax(eigvals)]
X_pca = X @ pc1

plt.scatter(X_pca, np.zeros_like(X_pca))
plt.title('Projected Data (1D PCA)')
plt.show()

# Sklearn PCA on Iris
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris = load_iris()
X_iris = iris.data
y_iris = iris.target

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_iris)

plt.scatter(X_reduced[:,0], X_reduced[:,1], c=y_iris)
plt.title('Iris PCA (2D)')
plt.show()
```

## Q3 – LDA (Manual + Sklearn)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Select 2 classes
X_bin = X_iris[y_iris != 2]
y_bin = y_iris[y_iris != 2]

# Class means
mean1 = np.mean(X_bin[y_bin==0], axis=0)
mean2 = np.mean(X_bin[y_bin==1], axis=0)

# Within-class scatter
S_w = np.cov(X_bin[y_bin==0].T) + np.cov(X_bin[y_bin==1].T)

# Between-class scatter
mean_diff = (mean1 - mean2).reshape(-1,1)
S_b = mean_diff @ mean_diff.T

# Projection vector
w = np.linalg.inv(S_w) @ (mean1 - mean2)

print('LDA Projection Vector:', w)

# Sklearn LDA (3 classes)
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X_iris, y_iris)

plt.scatter(X_lda[:,0], X_lda[:,1], c=y_iris)
plt.title('LDA Projection (Iris)')
plt.show()
```
