## Q1 – Normal Distribution Heights

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

mean = 165
std = 10

prob_160_170 = stats.norm.cdf(170, mean, std) - stats.norm.cdf(160, mean, std)
prob_above_180 = 1 - stats.norm.cdf(180, mean, std)

print('P(160-170):', prob_160_170)
print('P(>180):', prob_above_180)

x = np.linspace(130, 200, 500)
plt.plot(x, stats.norm.pdf(x, mean, std))
plt.title('Normal Distribution Curve')
plt.show()
```

## Q2 – Exponential Distribution (Service Times)

```python
mean_service = 5
lambda_param = 1/mean_service

prob_less_3 = stats.expon.cdf(3, scale=mean_service)
prob_more_10 = 1 - stats.expon.cdf(10, scale=mean_service)

print('P(<3 mins):', prob_less_3)
print('P(>10 mins):', prob_more_10)

x = np.linspace(0, 20, 500)
plt.plot(x, stats.expon.pdf(x, scale=mean_service))
plt.title('Exponential Distribution Curve')
plt.show()
```

## Q3 – Normality Check (Q-Q Plot)

```python
data = np.random.normal(75, 10, 100)
stats.probplot(data, dist='norm', plot=plt)
plt.title('Q-Q Plot')
plt.show()
```

## Q7 – Bernoulli Distribution (Coin Toss)

```python
p = 0.5
mean = p
variance = p*(1-p)
print('Mean:', mean)
print('Variance:', variance)

x = [0,1]
pmf = [1-p, p]
plt.bar(x, pmf)
plt.title('Bernoulli Distribution')
plt.show()
```

## Q8 – Binomial Distribution (n=10, p=0.6)

```python
from scipy.stats import binom

n = 10
p = 0.6

prob_7 = binom.pmf(7, n, p)
prob_atleast_8 = 1 - binom.cdf(7, n, p)

print('P(X=7):', prob_7)
print('P(X>=8):', prob_atleast_8)

x = np.arange(0, n+1)
plt.bar(x, binom.pmf(x, n, p))
plt.title('Binomial Distribution')
plt.show()
```
