## Q1 – Poisson Distribution: Call Center

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

lambda_calls = 5

p_3 = stats.poisson.pmf(3, lambda_calls)
p_more_6 = 1 - stats.poisson.cdf(6, lambda_calls)

print('P(X=3):', p_3)
print('P(X>6):', p_more_6)

sim_calls = np.random.poisson(lambda_calls, 200)
plt.hist(sim_calls, bins=15)
plt.title('Simulated Call Distribution (200 hours)')
plt.show()
```

## Q2 – Poisson: Hospital Emergency Visits

```python
lambda_30min = 4

p_4 = stats.poisson.pmf(4, lambda_30min)
p_less_2 = stats.poisson.cdf(1, lambda_30min)

print('P(exactly 4 in 30 min):', p_4)
print('P(less than 2):', p_less_2)

sim_hospital = np.random.poisson(lambda_30min, 500)
plt.hist(sim_hospital, bins=15)
plt.title('Simulated Emergency Visits (500 trials)')
plt.show()
```

## Q3 – Z-Score Probability: Exam Marks

```python
mean_marks = 70
sd_marks = 10

z_85 = (85 - mean_marks) / sd_marks
p_above_85 = 1 - stats.norm.cdf(85, mean_marks, sd_marks)
p_60_80 = stats.norm.cdf(80, mean_marks, sd_marks) - stats.norm.cdf(60, mean_marks, sd_marks)

print('Z-score (85):', z_85)
print('P(score > 85):', p_above_85)
print('P(60 < score < 80):', p_60_80)
```

## Q4 – Z-Score Probability: Heights

```python
mean_h = 165
sd_h = 8

z_180 = (180 - mean_h) / sd_h
p_taller_180 = 1 - stats.norm.cdf(180, mean_h, sd_h)
p_160_170 = stats.norm.cdf(170, mean_h, sd_h) - stats.norm.cdf(160, mean_h, sd_h)

print('Z-score (180cm):', z_180)
print('P(height > 180):', p_taller_180)
print('P(160 < height < 170):', p_160_170)
```

## Q5 – Normal Distribution: PDF & CDF

```python
mean_w = 50
sd_w = 5

pdf_55 = stats.norm.pdf(55, mean_w, sd_w)
cdf_55 = stats.norm.cdf(55, mean_w, sd_w)

print('PDF at 55:', pdf_55)
print('CDF at 55:', cdf_55)

x = np.linspace(30, 70, 400)
plt.plot(x, stats.norm.pdf(x, mean_w, sd_w))
plt.title('Normal PDF')
plt.show()

plt.plot(x, stats.norm.cdf(x, mean_w, sd_w))
plt.title('Normal CDF')
plt.show()
```

## Q6 – Exponential Distribution: PDF & CDF

```python
lambda_exp = 0.2
scale = 1/lambda_exp

pdf_5 = stats.expon.pdf(5, scale=scale)
cdf_5 = stats.expon.cdf(5, scale=scale)
p_gt_10 = 1 - stats.expon.cdf(10, scale=scale)

print('PDF at x=5:', pdf_5)
print('CDF at x=5:', cdf_5)
print('P(service time > 10):', p_gt_10)

x = np.linspace(0, 30, 400)
plt.plot(x, stats.expon.pdf(x, scale=scale))
plt.title('Exponential PDF')
plt.show()

plt.plot(x, stats.expon.cdf(x, scale=scale))
plt.title('Exponential CDF')
plt.show()
```
