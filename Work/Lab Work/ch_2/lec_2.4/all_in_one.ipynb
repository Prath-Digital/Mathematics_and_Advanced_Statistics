## Q1. Conditional Probability – Cards
- Deck: 52 cards
- Event A: First draw is a Heart
- Event B: Second draw is a Heart (without replacement)
- Compute P(B|A) and P(A ∩ B)

```python
total_cards = 52
hearts = 13

P_A = hearts / total_cards
P_B_given_A = (hearts - 1) / (total_cards - 1)
P_A_and_B = P_A * P_B_given_A

P_A, P_B_given_A, P_A_and_B
```

## Q4. Bayes’ Theorem – Medical Test
- Disease prevalence P(D) = 0.02
- Sensitivity P(+|D) = 0.95
- Specificity P(-|¬D) = 0.98
- Compute P(D|+) and P(¬D|+)

```python
P_D = 0.02
P_not_D = 1 - P_D
P_pos_given_D = 0.95
P_pos_given_not_D = 0.02

P_pos = (P_pos_given_D * P_D) + (P_pos_given_not_D * P_not_D)

P_D_given_pos = (P_pos_given_D * P_D) / P_pos
P_not_D_given_pos = 1 - P_D_given_pos

P_D_given_pos, P_not_D_given_pos
```

## Q5. Bayes’ Theorem – Email Spam Filter
- P(Spam) = 0.3
- P(Not Spam) = 0.7
- P(WIN | Spam) = 0.6
- P(WIN | Not Spam) = 0.05
- Compute P(Spam | WIN)

```python
P_spam = 0.3
P_not_spam = 0.7
P_WIN_given_spam = 0.6
P_WIN_given_not_spam = 0.05

P_WIN = (P_WIN_given_spam * P_spam) + (P_WIN_given_not_spam * P_not_spam)
P_spam_given_WIN = (P_WIN_given_spam * P_spam) / P_WIN

P_spam_given_WIN
```

## Q6. Bayes’ Theorem – Manufacturing Quality Control
- Machines: M1 = 50%, M2 = 30%, M3 = 20%
- Defect rates:
  - P(D|M1) = 0.01
  - P(D|M2) = 0.03
  - P(D|M3) = 0.05
- Product is defective
- Compute P(M1|D), P(M2|D), P(M3|D)

```python
P_M1, P_M2, P_M3 = 0.5, 0.3, 0.2
P_D_given_M1, P_D_given_M2, P_D_given_M3 = 0.01, 0.03, 0.05

P_D = (
    P_D_given_M1 * P_M1 +
    P_D_given_M2 * P_M2 +
    P_D_given_M3 * P_M3
)

P_M1_given_D = (P_D_given_M1 * P_M1) / P_D
P_M2_given_D = (P_D_given_M2 * P_M2) / P_D
P_M3_given_D = (P_D_given_M3 * P_M3) / P_D

P_M1_given_D, P_M2_given_D, P_M3_given_D
```
