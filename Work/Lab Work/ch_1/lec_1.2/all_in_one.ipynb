## Q1
- Explain Descriptive Statistics
- Explain Inferential Statistics
- Differentiate between the two

**Descriptive Statistics:**
- Summarizes and describes the main features of a dataset.
- Uses measures like mean, median, mode, variance, and standard deviation.
- Does not make predictions beyond the given data.

**Inferential Statistics:**
- Uses sample data to make predictions or inferences about a population.
- Uses probability, hypothesis testing, confidence intervals.
- Helps in decision-making under uncertainty.

**Difference (Mind Map Style):**
- Descriptive → What the data shows
- Inferential → What the data suggests about a population

## Q2
- Load a sample dataset
- Use describe()
- Explain each statistic

```python
import pandas as pd

data = {
    'Scores': [55, 60, 65, 70, 75, 80, 85, 90]
}

df = pd.DataFrame(data)
df
```

```python
df.describe()
```

**Explanation of describe():**
- count → number of observations
- mean → average value
- std → standard deviation (spread of data)
- min → minimum value
- 25%, 50%, 75% → quartiles
- max → maximum value

## Q3
- Mean using pandas and numpy
- Median using pandas and numpy
- Mode using pandas
- Bar chart comparison

```python
import numpy as np

mean_pd = df['Scores'].mean()
mean_np = np.mean(df['Scores'])

median_pd = df['Scores'].median()
median_np = np.median(df['Scores'])

mode_pd = df['Scores'].mode()[0]

mean_pd, mean_np, median_pd, median_np, mode_pd
```

```python
import matplotlib.pyplot as plt

values = [mean_pd, median_pd, mode_pd]
labels = ['Mean', 'Median', 'Mode']

plt.bar(labels, values)
plt.title('Mean vs Median vs Mode')
plt.ylabel('Value')
plt.show()
```

The bar chart shows how mean, median, and mode compare for the dataset.

## Q4
- Range
- Variance (pandas & numpy)
- Standard deviation
- Boxplot interpretation

```python
data_range = df['Scores'].max() - df['Scores'].min()

variance_pd = df['Scores'].var()
variance_np = np.var(df['Scores'], ddof=1)

std_pd = df['Scores'].std()
std_np = np.std(df['Scores'], ddof=1)

data_range, variance_pd, variance_np, std_pd, std_np
```

```python
plt.boxplot(df['Scores'])
plt.title('Boxplot of Scores')
plt.ylabel('Scores')
plt.show()
```

**Interpretation:**
- The boxplot shows data spread and median clearly.
- No extreme outliers are present.
- Dispersion appears moderate.
