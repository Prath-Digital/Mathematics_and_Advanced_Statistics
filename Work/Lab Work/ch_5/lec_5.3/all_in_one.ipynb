## Q1 – Angle Between Two Vectors

```python
import numpy as np

a = np.array([1, 2])
b = np.array([3, 4])

cos_theta = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
theta = np.degrees(np.arccos(cos_theta))

print('Angle (degrees):', theta)

# Random 3D vectors
v1 = np.random.rand(3)
v2 = np.random.rand(3)

angle = np.degrees(np.arccos(np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))))
print('Random 3D Angle:', angle)
```

## Q2 – Vector Projection

```python
import matplotlib.pyplot as plt

a = np.array([3, 4])
b = np.array([1, 2])

proj = (np.dot(a, b) / np.dot(b, b)) * b

print('Projection of a on b:', proj)

# Visualization
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, label='b')
plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, label='proj')
plt.legend()
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.title('Vector Projection')
plt.show()
```

## Q3 – Line in 2D

## Q4 – Plane in 3D

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
x, y = np.meshgrid(x, y)

z = 5 - x - y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
plt.title('Plane: x + y + z = 5')
plt.show()
```

## Q7 – Determinant & Inverse

```python
M = np.random.rand(3,3)

det = np.linalg.det(M)
inv = np.linalg.inv(M)

print('Matrix:\n', M)
print('Determinant:', det)
print('Inverse:\n', inv)
```

## Q8 – Verify A * A⁻¹ = I

```python
identity = np.dot(M, inv)
print('A * A⁻¹ ≈ I:\n', identity)
```
