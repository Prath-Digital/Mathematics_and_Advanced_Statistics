## Q1 – Euclidean Distance (Manual + SciPy)

```python
import numpy as np
from scipy.spatial.distance import euclidean

A = np.array([2, 3])
B = np.array([7, 8])

manual_dist = np.sqrt(np.sum((A - B)**2))
scipy_dist = euclidean(A, B)

print('Manual Distance:', manual_dist)
print('SciPy Distance:', scipy_dist)
```

## Q2 – Pairwise Distance Matrix

```python
from scipy.spatial.distance import cdist

points = np.random.rand(10, 2)
distance_matrix = cdist(points, points)

print('Distance Matrix:\n', distance_matrix)
```

## Q3 – Vector Norms

```python
v = np.array([4, -3, 7])

l1 = np.sum(np.abs(v))
l2 = np.sqrt(np.sum(v**2))

np_l1 = np.linalg.norm(v, 1)
np_l2 = np.linalg.norm(v, 2)

print('Manual L1:', l1)
print('Manual L2:', l2)
print('NumPy L1:', np_l1)
print('NumPy L2:', np_l2)
```

## Q4 – L1 vs L2 (Random 5D Vector)

```python
vec = np.random.randint(-10, 10, 5)

l1 = np.linalg.norm(vec, 1)
l2 = np.linalg.norm(vec, 2)

print('Vector:', vec)
print('L1 Norm:', l1)
print('L2 Norm:', l2)
```

## Q5 – Unit Vector Function

```python
def unit_vector(v):
    norm = np.linalg.norm(v)
    return v / norm

v = np.array([3, 4])
print('Unit Vector:', unit_vector(v))
```

## Q6 – Normalize Multiple Vectors

```python
vectors = np.random.rand(5, 3)

for v in vectors:
    unit_v = v / np.linalg.norm(v)
    print('Norm:', np.linalg.norm(unit_v))
```

## Q7 – Dot Product

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

manual_dot = np.sum(a * b)
numpy_dot = np.dot(a, b)

print('Manual Dot:', manual_dot)
print('NumPy Dot:', numpy_dot)
```

## Q8 – Orthogonality Check

```python
v1 = np.random.rand(3)
v2 = np.random.rand(3)

dot = np.dot(v1, v2)

print('Dot Product:', dot)
print('Orthogonal:', np.isclose(dot, 0))
```

## Q9 – Cross Product

```python
u = np.array([2, 3, 4])
v = np.array([5, 6, 7])

manual_cross = np.array([
    u[1]*v[2] - u[2]*v[1],
    u[2]*v[0] - u[0]*v[2],
    u[0]*v[1] - u[1]*v[0]
])

numpy_cross = np.cross(u, v)

print('Manual Cross:', manual_cross)
print('NumPy Cross:', numpy_cross)
```

## Q10 – Right-Handed System Check

```python
A = np.random.rand(3)
B = np.random.rand(3)
C = np.random.rand(3)

matrix = np.vstack([A, B, C])
det = np.linalg.det(matrix)

print('Determinant:', det)
print('Right-Handed System:', det > 0)
```
