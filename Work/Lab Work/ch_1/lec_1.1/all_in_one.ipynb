## Q3
- Create DataFrame with Name, Age, Department, Salary
- Show data types
- Filter numerical and categorical columns

```python
data = {
    'Name': ['Amit', 'Neha', 'Raj', 'Priya'],
    'Age': [28, 32, 26, 30],
    'Department': ['IT', 'HR', 'IT', 'Finance'],
    'Salary': [60000, 65000, 58000, 70000]
}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
numerical = df.select_dtypes(include='number')
categorical = df.select_dtypes(include='object')

numerical, categorical
```

## Q4
- Visualizations
- Describe patterns

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.histplot(df['Age'], bins=5)
plt.title('Age Distribution')
plt.show()
```

The age distribution shows most employees are between 26 and 32 years.

```python
sns.countplot(x='Department', data=df)
plt.title('Department Count')
plt.show()
```

IT department has the highest number of employees.

```python
df['Department'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Department Percentage')
plt.ylabel('')
plt.show()
```

The pie chart shows IT dominates the workforce distribution.

```python
sns.boxplot(y='Salary', data=df)
plt.title('Salary Distribution')
plt.show()
```

Salary distribution shows moderate variation with no extreme outliers.

## Q5
- Population vs Sample statistics

```python
import numpy as np

population = np.random.randint(40, 100, 1000)
sample = np.random.choice(population, 100)

population.mean(), population.std(), sample.mean(), sample.std()
```

**Difference:**
- Population statistics describe the entire dataset.
- Sample statistics estimate population characteristics and may vary.
