## Q1 – Eigenvalues & Eigenvectors

```python
import numpy as np

A = np.array([[4, 2], [1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print('Matrix A:\n', A)
print('Eigenvalues:', eigenvalues)
print('Eigenvectors:\n', eigenvectors)
```

## Q2 – Verify Av = λv

```python
# Take first eigenpair
v = eigenvectors[:, 0]
lam = eigenvalues[0]

Av = np.dot(A, v)
lv = lam * v

print('Av:', Av)
print('λv:', lv)
```

## Q3 – Random 3x3 Eigen Decomposition

```python
M = np.random.rand(3,3)

eigvals, eigvecs = np.linalg.eig(M)

print('Matrix:\n', M)
print('Eigenvalues:', eigvals)
print('Eigenvectors:\n', eigvecs)
```

## Q4 – LU Decomposition

```python
from scipy.linalg import lu

P, L, U = lu(M)

print('P:\n', P)
print('L:\n', L)
print('U:\n', U)

# Verify
print('Reconstructed:\n', P @ L @ U)
```

## Q5 – QR Decomposition

```python
Q, R = np.linalg.qr(M)

print('Q:\n', Q)
print('R:\n', R)

# Verify A = QR
print('QR Reconstruction:\n', Q @ R)
```
