## Q1 – Salary Data (Log-Normal Distribution)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

mu = 10.5
sigma = 0.4

prob_gt_60000 = 1 - stats.lognorm.cdf(60000, s=sigma, scale=np.exp(mu))
prob_30_50 = stats.lognorm.cdf(50000, s=sigma, scale=np.exp(mu)) - stats.lognorm.cdf(30000, s=sigma, scale=np.exp(mu))

print('P(>60000):', prob_gt_60000)
print('P(30000-50000):', prob_30_50)

x = np.linspace(10000, 150000, 500)
plt.plot(x, stats.lognorm.pdf(x, s=sigma, scale=np.exp(mu)))
plt.title('Log-Normal Salary Distribution')
plt.show()
```

## Q2 – Stock Returns (Log-Normal)

```python
mu_r = 0.02
sigma_r = 0.1

prob_gt_10 = 1 - stats.norm.cdf(0.10, mu_r, sigma_r)
print('P(return > 10%):', prob_gt_10)

returns = np.random.normal(mu_r, sigma_r, 100)
plt.hist(returns, bins=20)
plt.title('Simulated Returns Histogram')
plt.show()
```

## Q3 – City Population (Power-Law)

```python
alpha = 2.5
xmin = 1000

prob_gt_100k = (100000/xmin)**(-alpha+1)
print('P(city > 100000):', prob_gt_100k)

samples = (np.random.pareto(alpha, 500) + 1) * xmin
plt.hist(samples, bins=50)
plt.xscale('log')
plt.yscale('log')
plt.title('Power-Law Distribution (Log-Log Scale)')
plt.show()
```

## Q4 – Social Network Connections (Power-Law)

```python
alpha_net = 3
friends = (np.random.pareto(alpha_net, 2000) + 1)

plt.hist(friends, bins=50)
plt.xscale('log')
plt.yscale('log')
plt.title('Social Network Degree Distribution')
plt.show()
```

## Q5 – Skewed Income Data (Box-Cox Transformation)

```python
income = np.random.exponential(50000, 1000)

transformed, lambda_val = stats.boxcox(income)
print('Optimal Lambda:', lambda_val)

plt.hist(income, bins=30)
plt.title('Original Skewed Income')
plt.show()

plt.hist(transformed, bins=30)
plt.title('After Box-Cox Transformation')
plt.show()
```

## Q6 – Hospital Waiting Times (Box-Cox + Q-Q Plot)

```python
waiting = np.random.exponential(30, 500)

trans_wait, lambda_wait = stats.boxcox(waiting)
print('Optimal Lambda:', lambda_wait)

stats.probplot(waiting, dist='norm', plot=plt)
plt.title('Q-Q Plot Before Transformation')
plt.show()

stats.probplot(trans_wait, dist='norm', plot=plt)
plt.title('Q-Q Plot After Box-Cox')
plt.show()
```
