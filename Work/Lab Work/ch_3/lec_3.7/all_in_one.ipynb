## Q1 – Covariance: Study Hours vs Exam Score

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

hours = np.random.uniform(1, 10, 20)
scores = 50 + 5*hours + np.random.normal(0, 5, 20)

cov_matrix = np.cov(hours, scores)
cov_value = cov_matrix[0,1]

print('Covariance:', cov_value)

plt.scatter(hours, scores)
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score')
plt.show()
```

## Q2 – Covariance: Advertising Budget vs Sales

```python
budget = np.random.uniform(1000, 10000, 30)
sales = 20000 + 3*budget + np.random.normal(0, 10000, 30)

cov_matrix2 = np.cov(budget, sales)
cov_value2 = cov_matrix2[0,1]

print('Covariance:', cov_value2)

plt.scatter(budget, sales)
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.title('Advertising Budget vs Sales')
plt.show()
```

## Q3 – Pearson Correlation: Height vs Weight

```python
height = np.random.normal(170, 10, 25)
weight = 0.5*height - 30 + np.random.normal(0, 5, 25)

r, p = stats.pearsonr(height, weight)
print('Pearson r:', r)
print('p-value:', p)

plt.scatter(height, weight)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')
plt.show()
```

## Q4 – Pearson Correlation: Temperature vs Ice Cream Sales

```python
temp = np.random.uniform(20, 40, 30)
ice_cream = 100 + 10*temp + np.random.normal(0, 20, 30)

r2, p2 = stats.pearsonr(temp, ice_cream)
print('Pearson r:', r2)
print('p-value:', p2)

plt.scatter(temp, ice_cream)
plt.xlabel('Temperature')
plt.ylabel('Ice Cream Sales')
plt.title('Temperature vs Ice Cream Sales')
plt.show()
```

## Q5 – Spearman Correlation: Academic Rank vs Sports Rank

```python
academic_rank = np.arange(1, 16)
sports_rank = np.random.permutation(academic_rank)

rho, p_s = stats.spearmanr(academic_rank, sports_rank)
print('Spearman rho:', rho)
print('p-value:', p_s)
```
