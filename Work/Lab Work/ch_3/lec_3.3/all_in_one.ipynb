## Q1. One-Sample Z-Test (Weights)
Objective: Check if sample mean differs from population mean (μ = 70 kg, σ = 8 kg).
- Generate sample of 50 weights.
- Perform one-sample z-test.
- Print z-value and p-value.

```python
import numpy as np
from scipy import stats

np.random.seed(42)

# Given
mu = 70
sigma = 8
n = 50

# Generate sample
sample = np.random.normal(loc=72, scale=8, size=n)
sample_mean = np.mean(sample)

# Z-score
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

sample_mean, z, p_value
```

## Q2. Two-Sample Z-Test (Male vs Female Heights)
Objective: Compare means of two independent samples.
- Known population standard deviations.
- Perform two-sample z-test.
- Print z-value and p-value.

```python
# Given population standard deviations
sigma1 = 7
sigma2 = 6

# Generate samples
male = np.random.normal(170, sigma1, 60)
female = np.random.normal(165, sigma2, 55)

mean1 = np.mean(male)
mean2 = np.mean(female)

# Z-score
z = (mean1 - mean2) / np.sqrt((sigma1**2 / len(male)) + (sigma2**2 / len(female)))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

mean1, mean2, z, p_value
```

## Q3. One-Sample Z-Test (Water Intake)
Objective: Test if average daily water intake differs from 2 liters.
- μ = 2 liters
- σ = 0.4
- n = 40

```python
mu = 2
sigma = 0.4
n = 40

# Generate sample
water = np.random.normal(2.1, sigma, n)
mean_water = np.mean(water)

# Z-score
z = (mean_water - mu) / (sigma / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z)))

mean_water, z, p_value
```

## Q4. Two-Sample Z-Test (Batch A vs Batch B Scores)
Objective: Compare average scores of two batches.
- Known population standard deviations.
- Perform two-sample z-test.
- Interpret result.

```python
# Given standard deviations
sigma_A = 10
sigma_B = 12

# Generate datasets
batch_A = np.random.normal(75, sigma_A, 45)
batch_B = np.random.normal(80, sigma_B, 50)

mean_A = np.mean(batch_A)
mean_B = np.mean(batch_B)

# Z-test
z = (mean_A - mean_B) / np.sqrt((sigma_A**2 / len(batch_A)) + (sigma_B**2 / len(batch_B)))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Mean A:", mean_A)
print("Mean B:", mean_B)
print("Z-score:", z)
print("P-value:", p_value)

if p_value < 0.05:
    print("Result: Significant difference between batches.")
else:
    print("Result: No significant difference between batches.")
```
