## Q1. Population vs Sample (Heights)
- Generate 10,000 heights (cm).
- Take sample of 100.
- Compare means.
- Plot histogram and show both means.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Population
population = np.random.normal(loc=170, scale=10, size=10000)
sample = np.random.choice(population, 100)

pop_mean = np.mean(population)
sample_mean = np.mean(sample)

plt.hist(population, bins=50, alpha=0.5)
plt.axvline(pop_mean)
plt.axvline(sample_mean)
plt.title("Population vs Sample Means")
plt.show()

pop_mean, sample_mean
```

## Q2. Students Test Scores & Passing Mark
- Create dataset of scores.
- Assume passing mark = 60.
- Compare how many score around/below 60.

```python
scores = np.random.normal(65, 10, 200)
passing_mark = 60

below_pass = np.sum(scores < passing_mark)
above_pass = np.sum(scores >= passing_mark)

below_pass, above_pass
```

## Q3. Daily Steps Dataset
- Create dataset of 5000 daily steps.
- Sample 50 users.
- Compare means.
- Plot both distributions.

```python
steps_population = np.random.normal(8000, 1500, 5000)
steps_sample = np.random.choice(steps_population, 50)

pop_steps_mean = np.mean(steps_population)
sample_steps_mean = np.mean(steps_sample)

plt.hist(steps_population, bins=40, alpha=0.5)
plt.hist(steps_sample, bins=20, alpha=0.7)
plt.title("Steps: Population vs Sample")
plt.show()

pop_steps_mean, sample_steps_mean
```

## Q4. Compare Two Groups (Test Scores)
- Simulate scores for Group A and B.

## Q5. Business Assumption (Ice Cream Sales)
