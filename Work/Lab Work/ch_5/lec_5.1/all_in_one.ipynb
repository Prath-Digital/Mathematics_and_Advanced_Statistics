## Q1 – Vector Representation

```python
import numpy as np

# Student marks vector
marks = np.array([85, 90, 78])  # Math, Science, English

# Dataset sample vector
sample = np.array([25, 50000, 3])  # age, salary, experience

print('Marks Vector:', marks)
print('Shape:', marks.shape, 'Dimensions:', marks.ndim)

print('Sample Vector:', sample)
print('Shape:', sample.shape, 'Dimensions:', sample.ndim)
```

## Q2 – 2D & 3D Representation

```python
import matplotlib.pyplot as plt

p1 = np.array([2, 3])
p2 = np.array([5, 7])

plt.scatter(*p1, label='Point (2,3)')
plt.scatter(*p2, label='Point (5,7)')
plt.legend()
plt.title('2D Points')
plt.grid()
plt.show()

# 3D example
p3d = np.array([2, 3, 5])
print('3D Point:', p3d)
```

## Q3 – Addition & Subtraction

```python
v1 = np.array([2, 3])
v2 = np.array([4, 1])

print('Addition:', v1 + v2)
print('Subtraction:', v1 - v2)
```

## Q4 – Scalar Multiplication

```python
v = np.array([2, 3])
print('Scalar Multiplication:', 3 * v)
```

## Q5 – Magnitude of Vector

```python
v = np.array([3, 4])
magnitude = np.linalg.norm(v)
print('Magnitude:', magnitude)
```

## Q6 – Distance Between Two Points

```python
p1 = np.array([2, 3])
p2 = np.array([5, 7])

distance = np.linalg.norm(p1 - p2)
print('Distance:', distance)
```

## Q7 – Random Points & Distances

```python
points = np.random.rand(5, 2)

print('Points:\n', points)

from scipy.spatial.distance import cdist
dist_matrix = cdist(points, points)

print('Distance Matrix:\n', dist_matrix)

plt.scatter(points[:,0], points[:,1])
plt.title('Random 2D Points')
plt.grid()
plt.show()
```
