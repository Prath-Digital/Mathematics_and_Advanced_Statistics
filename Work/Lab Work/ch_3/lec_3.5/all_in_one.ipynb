## Q1 – Pearson Correlation

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
np.random.seed(42)

hours = np.random.uniform(1, 10, 30)
scores = 50 + 5*hours + np.random.normal(0, 5, 30)

r, p = stats.pearsonr(hours, scores)
print('Correlation (r):', r)
print('p-value:', p)

plt.scatter(hours, scores)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Correlation Plot')
plt.show()
```

## Q2 – Simple Linear Regression

```python
X = sm.add_constant(hours)
model = sm.OLS(scores, X).fit()
print(model.summary())
```

## Q3 – Multiple Linear Regression

```python
experience = np.random.randint(1, 20, 50)
education = np.random.randint(1, 4, 50)
salary = 20000 + 3000*experience + 5000*education + np.random.normal(0, 5000, 50)

X_multi = np.column_stack((experience, education))
X_multi = sm.add_constant(X_multi)
model_multi = sm.OLS(salary, X_multi).fit()
print(model_multi.summary())
```

## Q4 – One-Way ANOVA

```python
group1 = np.random.normal(70, 5, 30)
group2 = np.random.normal(75, 5, 30)
group3 = np.random.normal(80, 5, 30)

f_stat, p_val = stats.f_oneway(group1, group2, group3)
print('F-statistic:', f_stat)
print('p-value:', p_val)
```

## Q5 – Two-Way ANOVA

```python
import pandas as pd
import statsmodels.formula.api as smf

data = pd.DataFrame({
    'Score': np.random.normal(75, 10, 60),
    'Gender': np.repeat(['Male', 'Female'], 30),
    'Method': np.tile(np.repeat(['A', 'B'], 15), 2)
})

model_tw = smf.ols('Score ~ C(Gender) + C(Method) + C(Gender):C(Method)', data=data).fit()
anova_table = sm.stats.anova_lm(model_tw, typ=2)
print(anova_table)
```

## Q6 – Mann-Whitney U Test

```python
groupA = np.random.exponential(50000, 30)
groupB = np.random.exponential(52000, 30)

u_stat, p_val = stats.mannwhitneyu(groupA, groupB)
print('U-statistic:', u_stat)
print('p-value:', p_val)
```

## Q7 – Wilcoxon Signed-Rank Test

```python
before = np.random.normal(80, 5, 25)
after = before - np.random.normal(2, 2, 25)

w_stat, p_val = stats.wilcoxon(before, after)
print('Wilcoxon statistic:', w_stat)
print('p-value:', p_val)
```
