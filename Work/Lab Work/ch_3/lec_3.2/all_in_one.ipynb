## Q1. Sample Mean, Standard Deviation & 95% Confidence Interval
- Generate sample of 100 heights.
- Compute mean and standard deviation.
- Compute 95% confidence interval.
- Plot sample with CI.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)

# Generate sample
sample = np.random.normal(loc=170, scale=8, size=100)

mean = np.mean(sample)
std = np.std(sample, ddof=1)

# 95% CI
confidence = 0.95
z = stats.norm.ppf((1 + confidence) / 2)
margin_error = z * (std / np.sqrt(len(sample)))

ci_lower = mean - margin_error
ci_upper = mean + margin_error

plt.hist(sample, bins=15)
plt.axvline(mean)
plt.axvline(ci_lower)
plt.axvline(ci_upper)
plt.title("Sample Heights with 95% Confidence Interval")
plt.show()

mean, std, (ci_lower, ci_upper)
```

## Q2. Normal Distribution Curve & Two-Tailed Shaded Area
- Plot normal distribution.
- Shade both tails.
- Explain shaded area represents p-value.

```python
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)

# Critical z value (example ±1.96)
z_crit = 1.96

plt.fill_between(x, y, where=(x <= -z_crit))
plt.fill_between(x, y, where=(x >= z_crit))

plt.title("Two-Tailed Test (Shaded = p-value region)")
plt.show()

print("The shaded regions represent the p-value area in a two-tailed hypothesis test.")
```

## Q3. Standard Normal Distribution & Critical Values
- Plot standard normal curve.
- Mark critical values for 90%, 95%, 99% confidence.

```python
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y)

z_90 = stats.norm.ppf(1 - 0.05)
z_95 = stats.norm.ppf(1 - 0.025)
z_99 = stats.norm.ppf(1 - 0.005)

plt.axvline(z_90)
plt.axvline(-z_90)
plt.axvline(z_95)
plt.axvline(-z_95)
plt.axvline(z_99)
plt.axvline(-z_99)

plt.title("Critical Values for 90%, 95%, 99% Confidence")
plt.show()

z_90, z_95, z_99
```
