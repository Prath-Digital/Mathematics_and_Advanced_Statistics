## Q1. One-Sample T-Test (Sleep Duration)
Objective: Check if average sleep differs from 8 hours.
- Sample size = 30
- Population mean μ = 8

```python
import numpy as np
from scipy import stats

np.random.seed(42)

# Generate sample (unknown population std)
sleep = np.random.normal(7.5, 1, 30)

# One-sample t-test
t_stat, p_value = stats.ttest_1samp(sleep, 8)

np.mean(sleep), t_stat, p_value
```

## Q2. Independent Two-Sample T-Test (Blood Pressure)
Objective: Compare smokers vs non-smokers.
- 40 values each group.
- Assume unequal variances (Welch's t-test).

```python
# Generate data
smokers = np.random.normal(140, 15, 40)
non_smokers = np.random.normal(130, 12, 40)

# Independent t-test (Welch)
t_stat, p_value = stats.ttest_ind(smokers, non_smokers, equal_var=False)

np.mean(smokers), np.mean(non_smokers), t_stat, p_value
```

## Q3. One-Sample T-Test (Teen Screen Time)
Objective: Check if screen time differs from 4 hours.
- Sample size = 25
- μ = 4

```python
# Generate sample
screen_time = np.random.normal(4.5, 1, 25)

# One-sample t-test
t_stat, p_value = stats.ttest_1samp(screen_time, 4)

np.mean(screen_time), t_stat, p_value
```

## Q4. Independent Two-Sample T-Test (Monthly Spending)
Objective: Compare spending under 30 vs above 30.
- 35 samples each group.
- Assume normal distribution.

```python
# Generate data
under_30 = np.random.normal(20000, 4000, 35)
above_30 = np.random.normal(25000, 5000, 35)

# Independent t-test
t_stat, p_value = stats.ttest_ind(under_30, above_30, equal_var=False)

print("Mean Under 30:", np.mean(under_30))
print("Mean Above 30:", np.mean(above_30))
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Result: Significant difference between age groups.")
else:
    print("Result: No significant difference between age groups.")
```
