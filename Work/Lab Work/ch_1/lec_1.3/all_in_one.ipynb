## Q1
- Generate a normally distributed dataset
- Plot histogram and KDE
- Overlay mean and standard deviation
- Plot bell curve (optional)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(loc=50, scale=10, size=1000)

mean = np.mean(data)
std = np.std(data)

sns.histplot(data, kde=True, stat='density', bins=30)
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mean, std))
plt.axvline(mean, linestyle='--')
plt.axvline(mean + std, linestyle=':')
plt.axvline(mean - std, linestyle=':')
plt.title('Normal Distribution with Mean and Std Dev')
plt.show()
```

## Q2
- Calculate 25th, 50th, and 75th percentiles
- Plot percentile lines

```python
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

q1, q2, q3
```

## Q4
- Generate two normal distributions
- Compare KDE plots

```python
data1 = np.random.normal(50, 5, 1000)
data2 = np.random.normal(60, 15, 1000)

sns.kdeplot(data1, label='Mean 50, Std 5')
sns.kdeplot(data2, label='Mean 60, Std 15')
plt.legend()
plt.title('Comparison of Two Normal Distributions')
plt.show()

np.mean(data1), np.std(data1), np.mean(data2), np.std(data2)
```

## Q6
- Compare normal and peaked (Laplace) distributions
- Calculate kurtosis
- Interpret kurtosis

```python
from scipy.stats import kurtosis

normal_data = np.random.normal(0, 1, 1000)
laplace_data = np.random.laplace(0, 1, 1000)

sns.kdeplot(normal_data, label='Normal')
sns.kdeplot(laplace_data, label='Laplace')
plt.legend()
plt.title('KDE Comparison for Kurtosis')
plt.show()

kurt_normal = kurtosis(normal_data)
kurt_laplace = kurtosis(laplace_data)

kurt_normal, kurt_laplace
```

**Interpretation:**
- Normal distribution → Mesokurtic
- Laplace distribution → Leptokurtic (more peaked)
