## Q1. Venn Diagram – Tea and Coffee Preference
- 40 students like tea
- 50 students like coffee
- 20 students like both

```python
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

venn2(subsets=(20, 30, 20), set_labels=('Tea', 'Coffee'))
plt.title('Tea vs Coffee Preference')
plt.show()
```

## Q2. Venn Diagram – Facebook and Instagram Users
- 60 use Facebook
- 45 use Instagram
- 25 use both
- Find probabilities for each region

```python
total = 80
fb_only = 60 - 25
insta_only = 45 - 25
both = 25

print('P(Facebook only):', fb_only / total)
print('P(Instagram only):', insta_only / total)
print('P(Both):', both / total)

venn2(subsets=(fb_only, insta_only, both), set_labels=('Facebook', 'Instagram'))
plt.title('Facebook vs Instagram Users')
plt.show()
```

## Q3. Contingency Table – Gender vs Tea/Coffee Preference

```python
import pandas as pd

data = {
    'Gender': ['Male','Female','Male','Female','Male','Female','Male','Female','Male','Female'],
    'Preference': ['Tea','Coffee','Coffee','Tea','Tea','Tea','Coffee','Coffee','Tea','Coffee']
}

df = pd.DataFrame(data)
contingency = pd.crosstab(df['Gender'], df['Preference'])
contingency
```

## Q5. Marginal Probability – Sum of Two Dice

```python
import itertools

dice = list(itertools.product(range(1,7), repeat=2))
sums = [a+b for a,b in dice]
df = pd.DataFrame({'Sum': sums})
df['Sum'].value_counts(normalize=True).sort_index()
```

## Q6. Marginal Probability – Pass vs Fail

```python
data = {
    'Study_Habit': ['Good','Poor','Good','Average','Poor','Good','Average','Poor'],
    'Result': ['Pass','Fail','Pass','Pass','Fail','Pass','Fail','Fail']
}

df = pd.DataFrame(data)
table = pd.crosstab(df['Study_Habit'], df['Result'])

print('Marginal Probability of Pass:', table['Pass'].sum() / table.values.sum())
print('Marginal Probability of Fail:', table['Fail'].sum() / table.values.sum())
```
